Libreria# Generación de Datos de Prueba

En este notebook se generan:

- 5,000 clientes
- 500 productos
- 100,000 ventas

Posteriormente se exportan a formato CSV para ser cargados en BigQuery.

```python
import pandas as pd
import random
from faker import Faker as Faker
from datetime import datetime, timedelta

fake = Faker()

random.seed(42)
Faker.seed(42)
```

Generar clientes:

```python
clientes = []

paises = [
    "Guatemala",
    "El Salvador",
    "Honduras",
    "Costa Rica"
]

for i in range(1, 5001):
    clientes.append({
        "id_cliente": i,
        "nombre": fake.name(),
        "pais": random.choice(paises)
    })

clientes_df = pd.DataFrame(clientes)

clientes_df.head()
```

Generar producto

```python
categorias = [
    "Electrónica",
    "Hogar",
    "Ropa",
    "Deportes",
    "Alimentos"
]

productos = []

for i in range(1, 501):
    productos.append({
        "id_producto": i,
        "producto": f"Producto_{i}",
        "categoria": random.choice(categorias)
    })

productos_df = pd.DataFrame(productos)

productos_df.head()
```

Generar Ventas

```python
ventas = []

sucursales = [
    "San Salvador",
    "Guatemala City",
    "Tegucigalpa",
    "San José"
]

fecha_inicio = datetime(2025, 1, 1)

for i in range(1, 100001):

    fecha = fecha_inicio + timedelta(
        days=random.randint(0, 365)
    )

    ventas.append({
        "id_venta": i,
        "fecha_venta": fecha.date(),
        "id_cliente": random.randint(1, 5000),
        "id_producto": random.randint(1, 500),
        "sucursal": random.choice(sucursales),
        "monto": round(random.uniform(10, 500), 2)
    })

ventas_df = pd.DataFrame(ventas)

ventas_df.head()
```

GUARDAR CSV

```python
clientes_df.to_csv("../datasets/clientes.csv", index=False)

productos_df.to_csv("../datasets/productos.csv", index=False)

ventas_df.to_csv("../datasets/ventas.csv", index=False)

print("Archivos CSV generados correctamente")
```
