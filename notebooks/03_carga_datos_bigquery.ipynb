# Parte 7 - Carga de Datos a BigQuery

```python
from google.cloud import bigquery

client = bigquery.Client.from_service_account_json(
    "../credentials/retail-analytics-lab-grupo2-a4bfa003951d.json"
)

print("Conexión exitosa")
```

LEER CSV

```python
import pandas as pd

clientes_df = pd.read_csv("../datasets/clientes.csv")
productos_df = pd.read_csv("../datasets/productos.csv")
ventas_df = pd.read_csv("../datasets/ventas.csv")
```

CARGAR CLIENTES

```python
tabla_id = "retail-analytics-lab-grupo2.retail_dw.clientes"

job = client.load_table_from_dataframe(
    clientes_df,
    tabla_id
)

job.result()

print("Clientes cargados")
```

CARGAR PRODUCTOS

```python
tabla_id = "retail-analytics-lab-grupo2.retail_dw.productos"

job = client.load_table_from_dataframe(
    productos_df,
    tabla_id
)

job.result()

print("Productos cargados")
```

CARGAR VENTAS

```python
from google.cloud import bigquery

schema = [
    bigquery.SchemaField("id_venta", "INT64"),
    bigquery.SchemaField("fecha_venta", "DATE"),
    bigquery.SchemaField("id_cliente", "INT64"),
    bigquery.SchemaField("id_producto", "INT64"),
    bigquery.SchemaField("sucursal", "STRING"),
    bigquery.SchemaField("monto", "FLOAT64"),
]

job_config = bigquery.LoadJobConfig(
    schema=schema,
    write_disposition="WRITE_TRUNCATE"
)

job = client.load_table_from_dataframe(
    ventas_df,
    "retail-analytics-lab-grupo2.retail_dw.ventas",
    job_config=job_config
)

job.result()

print("Ventas cargadas correctamente")
```

VERIFICACION

```python
query = """
SELECT COUNT(*) total
FROM retail_dw.ventas
"""

resultado = client.query(query).to_dataframe()

resultado
```
