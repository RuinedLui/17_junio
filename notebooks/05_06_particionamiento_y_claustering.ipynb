Paso 1 - Crear tabla particionada

Creamos una nueva tabla particionada utilizando la columna

```python
CREATE TABLE retail_dw.ventas_particionadas
(
    id_venta INT64,
    fecha_venta DATE,
    id_cliente INT64,
    id_producto INT64,
    sucursal STRING,
    monto NUMERIC
)
PARTITION BY fecha_venta;
```

Paso 4 - Consulta sobre tabla particionada

```python
SELECT sucursal, SUM(monto) as total_ventas
FROM `retail_dw.ventas_particionadas`
WHERE fecha_venta BETWEEN '2025-01-01' AND '2026-01-01' 
  AND sucursal = 'San Salvador'
GROUP BY sucursal;
```

Paso 1 - Crear tabla clusterizada

```python
CREATE TABLE retail_dw.ventas_cluster
(
    id_venta INT64,
    fecha_venta DATE,
    id_cliente INT64,
    id_producto INT64,
    sucursal STRING,
    monto NUMERIC
)
PARTITION BY fecha_venta
CLUSTER BY sucursal;
```

Consulta sobre tabla Clausterizada

```python
SELECT sucursal, SUM(monto) as total_ventas
FROM `retail_dw.ventas_cluster`
WHERE fecha_venta BETWEEN '2025-01-01' AND '2026-01-01' 
  AND sucursal = 'San Salvador'
GROUP BY sucursal;
```
