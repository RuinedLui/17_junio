```python
from google.cloud import bigquery

client = bigquery.Client.from_service_account_json(
    "../credentials/retail-analytics-lab-grupo2-a4bfa003951d.json"
)

print("Conexión exitosa")
```

```python
dataset_id = f"{client.project}.retail_dw"

dataset = bigquery.Dataset(dataset_id)
dataset.location = "US"

dataset = client.create_dataset(dataset, exists_ok=True)

print(f"Dataset creado: {dataset.dataset_id}")
```

```python
clientes_sql = """
CREATE TABLE IF NOT EXISTS retail_dw.clientes (
    id_cliente INT64,
    nombre STRING,
    pais STRING
);
"""

client.query(clientes_sql).result()

print("Tabla clientes creada")
```

```python
productos_sql = """
CREATE TABLE IF NOT EXISTS retail_dw.productos (
    id_producto INT64,
    producto STRING,
    categoria STRING
);
"""

client.query(productos_sql).result()

print("Tabla productos creada")
```

```python
ventas_sql = """
CREATE TABLE IF NOT EXISTS retail_dw.ventas (
    id_venta INT64,
    fecha_venta DATE,
    id_cliente INT64,
    id_producto INT64,
    sucursal STRING,
    monto NUMERIC
);
"""

client.query(ventas_sql).result()

print("Tabla ventas creada")
```

```python
tables = client.list_tables("retail_dw")

for table in tables:
    print(table.table_id)
```
